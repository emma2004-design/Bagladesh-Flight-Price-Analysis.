```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

plt.figure(figsize=(10,5))
sns.histplot(df["Base Fare (BDT)"], bins=50, kde=True, color="skyblue")
plt.xlabel("Base Fare (BDT)")
plt.ylabel("Frequency")
plt.title("Distribution of Base Fare (BDT)")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Airline", y="Base Fare (BDT)", data=df, estimator=np.mean, palette="coolwarm")
plt.xticks(rotation=90)
plt.xlabel("Airline")
plt.ylabel("Average Price")
plt.title("Average Base Fare (BDT) by Airline")
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.boxplot(x="Source", y="Base Fare (BDT)", hue="Destination", data=df)
plt.xticks(rotation=45)
plt.xlabel("Source City")
plt.ylabel("Base Fare (BDT)")
plt.title("Base Fare (BDT)s from Different Sources to Destinations")
plt.legend(loc="upper right")
plt.show()
```

```python
df["Departure Date & Time"] = pd.to_datetime(df["Departure Date & Time"], dayfirst=True)
```

```python
df["Departure_Day"] = df["Departure Date & Time"].dt.day
df["Departure_Month"] = df["Departure Date & Time"].dt.month
df["Departure_Weekday"] = df["Departure Date & Time"].dt.weekday
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Departure_Weekday", y="Base Fare (BDT)", data=df, estimator=np.mean, palette="coolwarm")
plt.xticks(ticks=[0,1,2,3,4,5,6], labels=["Mon","Tue","Wed","Thu","Fri","Sat","Sun"])
plt.xlabel("Day of the Week")
plt.ylabel("Average Base Fare (BDT)")
plt.title("Flight Base Fare (BDT) by Day of the Week")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="Departure_Month", y="Base Fare (BDT)", data=df, palette="viridis")
plt.xlabel("Month")
plt.ylabel("Base Fare (BDT)")
plt.title("Base Fare (BDT) by Month")
plt.show()
```

```python
df["Departure Date & Time"] = pd.to_datetime(df["Departure Date & Time"], errors='coerce')

# Extract useful features: Day, Month, Hour
df["Dep_Day"] = df["Departure Date & Time"].dt.day
df["Dep_Month"] = df["Departure Date & Time"].dt.month
df["Dep_Hour"] = df["Departure Date & Time"].dt.hour

df.head()  # Display the first few rows to verify
df["Dep_Day"] = df["Departure Date & Time"].dt.day
df["Dep_Month"] = df["Departure Date & Time"].dt.month
df["Dep_Hour"] = df["Departure Date & Time"].dt.hour

df.head()  # Display the first few rows to verify
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set up the canvas for 3 side-by-side bar plots
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

# Plot Average Flight Prices by Month, Day, and Hour
# Changed 'Price' to 'Base Fare (BDT)' assuming this is the correct column for price
sns.barplot(data=df, x="Dep_Month", y="Base Fare (BDT)", ax=axes[0])
axes[0].set_title("Average Flight Price by Month")

sns.barplot(data=df, x="Dep_Day", y="Base Fare (BDT)", ax=axes[1])
axes[1].set_title("Average Flight Price by Day")

sns.barplot(data=df, x="Dep_Hour", y="Base Fare (BDT)", ax=axes[2])
axes[2].set_title("Average Flight Price by Hour of Departure")

plt.tight_layout()
plt.show()
```
